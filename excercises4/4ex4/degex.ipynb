Return: An array D[1..n]

```python
dcount = [0] * 1000 
with open("rosalind_deg.txt", "r") as f:
    nodes, edges = map(int, f.readline().strip().split(" "))
    for line in f:
        u, v = map(int, line.strip().split(" "))
        dcount[u - 1] += 1 
        dcount[v - 1] += 1  


outputd = " ".join(map(str, dcount[:nodes])) 
print(outputd)
```
