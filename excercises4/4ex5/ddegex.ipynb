 vertices in the edge list format.

Return: An array D[1..n]
 where D[i]
 is the sum of the degrees of i
's neighbors

```python
with open("rosalind_ddeg.txt", "r") as f:
    n, edges = f.readline().strip().split(" ")
    edges_list = [line.strip().split(" ") for line in f]

nodes_neig = {str(i+1): [] for i in range(int(n))}

for u, v in edges_list:
    nodes_neig[u].append(v)
    nodes_neig[v].append(u)

for i in range(int(n)):
    totsum = 0
    currnode = str(i + 1)
    
    for neighbor in nodes_neig[currnode]:
        totsum += len(nodes_neig[neighbor]) 
    
    print(totsum, end=" ")
```
