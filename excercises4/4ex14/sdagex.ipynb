Given: A weighted DAG with integer edge weights from −103
 to 103
 and n≤105
 vertices in the edge list format.

Return: An array D[1..n]
 where D[i]
 is the length of a shortest path from the vertex 1
 to the vertex i
 (D[1]=0
). If i
 is not reachable from 1
 set D[i]
 to x

```python
with open("rosalind_sdag.txt", "r") as f:
    vertice, edge = map(int, f.readline().strip().split(" "))
    graph = {v:{} for v in range(1, vertice+1)}
    for line in f:
        vertice1, vertice2, weight = map(int, line.strip().split(" "))
        graph[vertice1][vertice2] = weight

def topsort(v, visited, stack):
    visited[v] = True
    for u in graph[v]:
        if not visited[u]:
            topsort(u, visited, stack)
    stack.append(v)

def path(graph, vertice, source):
    visited = [False] * (vertice + 1)
    stack = []
    for i in range(1, vertice + 1):
        if not visited[i]:
            topsort(i, visited, stack)
    INF = float('inf') 
    distance = [INF] * (vertice + 1)
    distance[source] = 0
    while stack:
        i = stack.pop()
        for u in graph[i]:
            if distance[i] != INF and distance[u] > distance[i] + graph[i][u]:
                distance[u] = distance[i] + graph[i][u]   
    return distance

distance = path(graph, vertice, 1)
for i in range(1, vertice + 1):  
    if distance[i] == float('inf'):
        print("x", end=" ")
    else:
        print(distance[i], end=" ")
```
