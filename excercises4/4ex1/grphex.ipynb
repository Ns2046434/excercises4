Given: A collection of DNA strings in FASTA format having total length at most 10 kbp.

Return: The adjacency list corresponding to O3
. You may return edges in any order

```python
seqs = {}
with open("rosalind_grph.txt", 'r') as fa:
    seq_id = ""
    seq = ""
    for line in fa:
        line = line.strip()
        if line.startswith(">"):
            if seq_id:  
                seqs[seq_id] = seq
                seq = ""
            seq_id = line[1:]
        else:
            seq += line 
    if seq_id: 
        seqs[seq_id] = seq

def overlap_graph(seqs, n=3):
    adlist = []
    seq_ids = list(seqs.keys())
    
    for i in range(len(seq_ids)):
        for j in range(len(seq_ids)):
            if i != j:
                seq1 = seqs[seq_ids[i]]
                seq2 = seqs[seq_ids[j]]
                if seq1.endswith(seq2[:n]): 
                    adlist.append((seq_ids[i], seq_ids[j]))
    
    return adlist

adlist = overlap_graph(seqs, n=3)
for edge in adlist:
    print(edge[0], edge[1])
```
